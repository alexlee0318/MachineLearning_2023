### 단순 선형회귀

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 아버지와 아들의 키 -Galton

```python
df = pd.read_csv('data/Galton.txt', sep='\t')
df.head()
```

- 아버지와 아들의 키만 발췌
- 컬럼명을 Father, Son으로 변경
- 단위(inch --> cm ) 변경

```python
# Filtering
df = df[df.Gender == 'M']
df.head(3)
```

```python
# Selection
df = df[['Father', 'Height']]
# Col명 변경
df.columns = ['Father', 'Son']
df.head(3)
```

```python
# 단위변경
df = df * 2.54
df.head(3)
```

- 산점도 구하기

```python
plt.figure(figsize=(5,5))
plt.scatter(df.Father, df.Son)
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.title('Father  vs  Son', fontsize=10)
plt.grid(), plt.show()
```

- 회귀선 구하고 그리기
    - np.linalg.lstsq()     # 최소 자승법(Least square)
    - np.polyfit()

```python
# 1차 식의 회귀식
weight, bias = np.polyfit(df.Father, df.Son, 1)     # X, Y, degree(차수)
weight, bias
```

```python
# 2차 식의 회귀식
np.polyfit(df.Father, df.Son, 2)
```

```python
# 직선 그리기
df.Father.max(), df.Father.min()
```

```python
# 회귀선(직선) 그리기
xs = np.array([157, 200])   # min, max 값
ys = xs * weight + bias
ys
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df.Father, df.Son)
plt.plot(xs, ys, c='red', lw=3)
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.title('Father  vs  Son', fontsize=10)
plt.grid(), plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.regplot(x='Father', y='Son', data=df)
plt.show()
```

- Scikit-learn 으로 회귀식 구하기

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.get_params()
```

```python
# X 데이터의 차원을 2로 만들어주어야 함
X = df.Father.values.reshape(-1, 1)
y = df.Son.values
X.shape, y.shape
```

```python
# 학습
lr.fit(X, y)
```

```python
# weight, bias
lr.coef_, lr.intercept_
# np.ployfit()으로 구한 값: (0.44774791....., 97.17763700.....) 과 동일.
```

- 잔차 제곱의 합(RSS: Residual Sum of Square)
- 평가 - R Squared

```python
lr.score(X, y)      # 1 -  Sum of Square Error(residue) / Sum of Square Total
```

```python
from sklearn.metrics import r2_score
pred = lr.predict(X)
r2_score(y, pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, pred)
```
