###단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

- 당뇨병 데이터셋 탐색

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()   # mean값이 0에 수렴한다(e-19에서 -19는 지수를 뜻함) // 분포가 N(0, 0.0476)로 이미 정규화 되어있음
```

```python
print(diabetes.DESCR)
```

- 10개의 feature와 target간의 상관관계도

```python
fig, axs = plt.subplots(2, 5, figsize=(16,8))
for i, feature in enumerate(df.columns[:-1]):
    row, col = i//5 , i%5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- feature(10개)  vs. target  __ 관계

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 각 feature와 target의 선형회귀를 했을 때의 R squared value
for feature in df.columns[:-1]:
    X= df[feature].values.reshape(-1, 1)  # 2차원식으로 reshape하기
    lr.fit(X, diabetes.target)
    score= lr.score(X, diabetes.target)
    print(f'{feature}:{score:.4f}') # bmi가 가장 좋음(높은수치)
```

#### BMI  vs target

- 훈련/테스트 데이터셋 분리(단, test_size=0.1 로)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2023
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 가중치(계수), bias(편향, 절편)
lr.coef_, lr.intercept_ # 950*bmi+151
```

```python
# R squared
lr.score(X_test, y_test)
```

```python
# R squared, MSE
from sklearn.metrics import r2_score, mean_squared_error
pred = lr.predict(X_test)
r2_score(y_test, pred), mean_squared_error(y_test, pred)    # R square값은 0~1 사이에서 움직임
# 4598이 뭔지 정확히 알 수 없지만 두 값을 넣었을 때 차이가 크게 날 것
```

- 시각화

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel('BMI'), plt.ylabel('target')
plt.title('BMI vs. target', fontsize=15)
plt.show()
```

```python
xs = np.array([-0.1, 0.18])
ys = np.dot(lr.coef_[0], xs) + lr.intercept_
ys
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, label='train dataset')
plt.plot(xs, ys, 'r-', lw=3)
plt.scatter(X_test, y_test, marker='^', label='test dataset')
plt.grid(), plt.legend()
plt.xlabel('BMI'), plt.ylabel('target')
plt.title('BMI vs. target', fontsize=15)
plt.show()
```

```python
r2_score(y_test2, pred2), mean_squared_error(y_test2, pred2)
    # squared error값이 커지는건 좋지 않은것 -> GsCV쓸 때는 scoring값이 커지므로 앞에 neg(-, negative뜻)를 붙임
```
