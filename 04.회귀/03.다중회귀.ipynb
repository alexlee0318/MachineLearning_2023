### 다중 선형회귀 - 보스톤 주택 가격

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
````

- 데이터 탐색

````python
df = pd.read_csv('data/Boston_house.csv')
df.shape
````

````python
df.head()
````

- Feature 별 산점도와 선형회귀선

````python
plt.figure(figsize=(4,3))
sns.regplot(x='AGE', y='Target',data=df);
````

````python
fig, axs = plt.subplots(3, 4, figsize=(16,12))
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i //4, i % 4
    sns.regplot(x=feature, y='Target', data=df, ax=axs[row][col])
````

- Feature 별 결정 계수(R squared)

````python
from sklearn.linear_model import LinearRegression
for feature in df.columns[:-1]:
    lr = LinearRegression()
    X = df[feature].values.reshape(-1, 1)
    lr.fit(X, df.Target.values)
    score = lr.score(X, df.Target.values)
    print(f'{feature}: {score:.4f}')    # LSTAT > RM > INDUS 순
````

#### 다중 선형 회귀 - feature를 모두 포함시켜서

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:,:-1].values, df.Target.values, test_size=0.1, random_state=2023   # reshape필요없음
)
````

````python
lr = LinearRegression()
lr.fit(X_train, y_train)
````

````python
# 결정 계수(R squared)
lr.score(X_test, y_test)
````

````python
# 가중치
lr.coef_
````

````python
# 편향
lr.intercept_
````

- 회귀식 구하는 공식 : ```Target= -0.0011 * AGE + 0.0076 * B + 3.9 * RM +...+ 2.9 * CHAS + 35.8```

- 테스트 데이터에 적용

````python
X_test[0], y_test[0]
````

````python
# 방법1) weight와 bias를 이용하여 직접 값 구하기
np.dot(lr.coef_, X_test[0]) + lr.intercept_
````

````python
# 방법2)  predict() method 사용
lr.predict(X_test[0].reshape(1, -1))[0]  # 두 방법 모두 결과 똑같음
````

````python
# 10개 테스트 데이터에 적용
pred = lr.predict(X_test)
for i in range(10):
    print(f'실제값: {y_test[i]},\t예측값: {pred[i]:.4f}')
````

```
모든 데이터에 적용했다고 값이 잘 나올까? -> (X)
다중회귀라고해서 모든 Feature에 적용할 필요 없음
방법 - 모든 Feature에서 꼴지를 제외하며 계산 or 일부 데이터부터 추가해가며 계산하는 방법 있음
```

````python
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
````

````python
svr = SVR()
svr.fit(X_train, y_train)
pred_sv = svr.predict(X_test)
````

````python
r2_sv = r2_score(y_test, pred_sv)
mse_sv = mean_squared_error(y_test, pred_sv)
````

- Random Forest Regressor

````python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=2023)
rfr.fit(X_train, y_train)
pred_rf = rfr.predict(X_test)
r2_rf = r2_score(y_test, pred_rf)
mse_rf = mean_squared_error(y_test, pred_rf)
````

- XGBoost

````python
from xgboost import XGBRegressor
xgr = XGBRegressor()
xgr.fit(X_train, y_train)
pred_xg = xgr.predict(X_test)
r2_xg = r2_score(y_test, pred_xg)
mse_xg = mean_squared_error(y_test, pred_xg)
````

- Linear Regression

````python
pred_lr = lr.predict(X_test)
r2_lr = r2_score(y_test, pred_lr)
mse_lr = mean_squared_error(y_test, pred_lr)
````

- 비교

````python
print('\t\tLR\tSVR\tRF\tXG')
print(f'R squared\t{r2_lr:.4f}\t{r2_sv:.4f}\t{r2_rf:.4f}\t{r2_xg:.4f}')
print(f'Mean squared E\t{mse_lr:.4f}\t{mse_sv:.4f}\t{mse_rf:.4f}\t{mse_xg:.4f}')
# XG성능이 좋음. 리니어말고 R리그레써도 쓸 수 있음.
````

````python
df = pd.DataFrame({'y':y_test, 'LR': pred_lr, 'SV': pred_sv, 'RF': pred_rf, 'XG': pred_xg})
df.head()   # 성능 좋음- 반드시 linear Regressor를 쓰지 않아도 됨
````

- 규제 선형회귀

-Ridge회귀 : 선형 회귀에 L2 규제 적용

````python
from sklearn.linear_model import Ridge
ridge = Ridge(random_state=2023)
ridge.get_params()
````

````python
ridge.fit(X_train, y_train)
pred_rg = ridge.predict(X_test)
r2_rg = r2_score(y_test, pred_rg)
mes_rg = mean_squared_error(y_test, pred_rg)
````

````python
r2_rg, mes_rg
````

- Lasso : 선형 회귀에 L1 규제 적용

````python
from sklearn.linear_model import Lasso
lasso = Lasso(random_state=2023)
lasso.get_params()
````

````python
lasso.fit(X_train, y_train)
pred_ls = lasso.predict(X_test)
r2_ls = r2_score(y_test, pred_ls)
mes_ls = mean_squared_error(y_test, pred_ls)
````

````python
r2_ls, mes_ls
````

````python
print('\t\tLR\tSV\tRF\tXG\tRidge\tLasso')
print(f'R squared\t{r2_lr:.4f}\t{r2_sv:.4f}\t{r2_rf:.4f}\t{r2_xg:.4f}\t{r2_rg:.4f}\t{r2_ls:.4f}')
print(f'Mean squared E\t{mse_lr:.4f}\t{mse_sv:.4f}\t{mse_rf:.4f}\t{mse_xg:.4f}\t{mes_rg:.4f}\t{mes_ls:.4f}')
````

````python
df['Ridge'] = pred_rg
df['Lasso'] = pred_ls
df.head()
````
