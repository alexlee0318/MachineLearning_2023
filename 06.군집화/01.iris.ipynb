### K-Means Algorithm - iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)

pca = PCA(2)
iris_pca = pca.fit_transform(iris_std)
```

#### 1. 군집화(Clustering)

```python
from sklearn.cluster import KMeans
KMeans().get_params()   # 기본세팅: 시작이 k-means++로, 300번 반복, 클러스터 8개, 시각화(verbose) 0
```

```python
kmeans = KMeans(n_clusters=3, random_state=2023, n_init='auto')  # iris가 3이니까
kmeans.fit(iris_std) #  y값 미지정-비지도학습이니까
```

```python
# 학습결과
kmeans.labels_
```

```python
np.unique(kmeans.labels_, return_counts=True)   
# 각 클러스터의 크기(수): 52, 50, 48
# 넘버링: 집단이 큰 순서대로 0부터 부여 
```

#### 2. 시각화

```python
df =pd.DataFrame(iris_pca, columns=['PC1','PC2'])
df['target'] = iris.target
df['cluster'] = kmeans.labels_
df.head(54).tail()  # cluster(labels_의 값)의 중간값을 보기위함
```

```python
# 데이터 분포
df.groupby(['target','cluster'])[['PC1']].count()
```

```python
markers = ['^','s','o']
plt.figure(figsize=(5,3))
for i, marker in enumerate(markers):
    x_data = df[df.cluster== i]['PC1']  # i = 0,1,2
    y_data = df[df.cluster== i]['PC2']
    plt.scatter(x_data, y_data, marker=marker, label='cluster'+str(i))
    plt.scatter(x_data.mean(), y_data.mean(), marker='*', c='red', s=100)   # 중심점 추가
plt.legend()
plt.xlabel('PC1'), plt.ylabel('PC2')
plt.title('Clustering result')
plt.show()
```

- PCA Original Data와 비교

```python
plt.figure(figsize=(10,3))
for k,column in enumerate (['target','cluster']):
    plt.subplot(1,2,k+1)
    for i , marker in enumerate(markers):
        x_data = df[df[column] == i]['PC1']
        y_data = df[df[column] == i]['PC2']
        if k == 0:
            plt.scatter(x_data, y_data, marker=marker, label=iris.target_names[i])
        else:
            plt.scatter(x_data, y_data, marker=marker, label='cluster'+str(i))
            plt.scatter(x_data.mean(), y_data.mean(), marker='*', c='red', s=100) # 중심점 추가

    plt.legend()
    plt.xlabel('PC1'), plt.ylabel('PC2')
    if k == 0:
        plt.title("Original data")
    else:
        plt.title("Clustered data")
plt.show()
```
